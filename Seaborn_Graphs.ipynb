```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
bss= sns.load_dataset("titanic")
sns.distplot(x= bss['age'] , bins=10)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
bss= sns.load_dataset("titanic")
sns.displot(x= bss['age'] , bins=10)
plt.show()
```
